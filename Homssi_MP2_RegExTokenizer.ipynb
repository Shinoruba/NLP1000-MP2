```

#pattern = r"\b[\w']+\b|https?://[a-zA-Z0-9./?=_-]+|\#\w+|@\w+|[!?.,:;]"        # test 1 AAAAAAA
pattern = r"https?://[^\s]+|#\w+|@\w+|\b[\w']+\b|[!?.,:;]"                      # test with numbers
```

```
# SAMPLE CODE SAMPLE CODE SAMPLE CODE
# Function to tokenize text
def sample_tokenize(text):
    return re.findall(pattern, text.lower())

# Test with sample corpus
sample_corpus = [
    "This is an example of a corpus for processing.",
    "There isn't much to it, but here it is.",
    "Also, here is a number, 55. Is this a word?"
]

all_sample_tokens = []
for sentence in sample_corpus:
    all_sample_tokens.extend(sample_tokenize(sentence))

sample_token_counts = Counter(all_sample_tokens)
sample_total_tokens = len(all_sample_tokens)
sample_total_vocabulary = len(sample_token_counts)

print("== Testing with sample corpus: ==")
print(f"Sample Total tokens: {sample_total_tokens}")
print(f"Sample Total vocabulary: {sample_total_vocabulary}")
print("All words + counts:")
for token, count in sample_token_counts.most_common():
    print(f"{token} : {count}")
```

```
# Function to tokenize text
def tokenize(text):
    return re.findall(pattern, text.lower())

# Tokenize all tweets
all_tokens = []
for tweet in tweets['text']:
    all_tokens.extend(tokenize(tweet))

# Count tokens and calculate vocabulary
token_counts = Counter(all_tokens)
total_tokens = len(all_tokens)
total_vocabulary = len(token_counts)
top_25 = token_counts.most_common(25)

# ============================================================
print("== Results for given tweet dataset: ==")
print(f"Total tokens: {total_tokens}")
print(f"Total vocabulary: {total_vocabulary}")
print("Top 25 tokens and their counts:")
for token, count in top_25:
    print(f"{token}: {count}")
```
